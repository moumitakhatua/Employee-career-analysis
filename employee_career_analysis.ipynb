```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
data=pd.read_csv("/bin/Your Career Aspirations of GenZ.csv")
```

```python
data.head()
```

ANALYSIS OF COUNTRY

```python
country=data['Your Current Country.'].value_counts()
label=country.index
count=country.values
colors=['blue','light green']
fig=go.Figure(data=[go.Pie(labels=label,values=count)])
fig.update_layout(title_text='current country')
fig.update_traces(hoverinfo='label+value',textinfo='percent',textfont_size=20,marker=dict(line=dict(color='black',width=2)))
fig.show()
```

WHAT ARE THE FACTORS INFLIUENCING CARRIER ASPIRATIONS OF GENZ ?

```python
ca=data['Which of the below factors influence the most about your career aspirations ?'].value_counts()
label=ca.index
count=ca.values
color=['blue','light pink','','aquatica','yellow']
fig=go.Figure(data=[go.Pie(labels=label,values=count)])
fig.update_layout(title_text='factors influence the most about your career aspirants')
fig.update_traces(hoverinfo='label+value',textinfo='percent',textfont_size=20,marker=dict(colors=color,line=dict(color='black',width=3)))
fig.show()
```

HOW MANY WANT TO PURSUE HIGHER EDUCATION OUTSIDE INDIA WITH THEIR INVESTMENT?

```python
q2=data['Would you definitely pursue a Higher Education / Post Graduation outside of India ? If only you have to self sponsor it.'].value_counts()
label=q2.index
count=q2.values
color=['blue','light green','yellow']
fig=go.Figure(data=[go.Pie(labels=label,values=count)])
fig.update_layout(title_text='education place')
fig.update_traces(hoverinfo='label+value',textinfo='percent',textfont_size=20,marker=dict(colors=color,line=dict(color='black',width=3)))
fig.show()
```

HOW LIKELY GENZ IS TO WORK FOR ONE COMPANY FOR THREE YEARS OR MORE ?

```python
q3=data['How likely is that you will work for one employer for 3 years or more ?'].value_counts()
label=q3.index
count=q3.values
color=['gold','aqua']
fig=go.Figure(data=[go.Pie(labels=label,values=count)])
fig.update_layout(title_text='work place')
fig.update_traces(hoverinfo='label+value',textinfo='percent',textfont_size=20,marker=dict(colors=color,line=dict(color='red',width=3)))
fig.show()
```

WHAT IS THE PREFERRED ENVIRONMENT OF GENZ ?

```python
q4=data['What is the most preferred working environment for you.'].value_counts()
label=q4.index
count=q4.values
color=['gold','aqua','magenta','blue','voilet','orange']
fig=go.Figure(data=[go.Pie(labels=label,values=count)])
fig.update_layout(title_text='working environment')
fig.update_traces(hoverinfo='label+value',textinfo='percent',textfont_size=30,marker=dict(colors=color,line=dict(color='black',width=3)))
fig.show()
```
